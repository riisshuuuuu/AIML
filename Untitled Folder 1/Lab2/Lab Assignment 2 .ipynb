## Data Cleaning

1. Analyse the data and identify which columns are not relevant for house rent prediction task. Drop 
those columns from the dataframes. 

```python
train=train.drop(['Posted On','Floor','Area Type','Area Locality','Tenant Preferred','Point of Contact'],axis=1)
train
```

2. Check for missing values and logically impute the dataset

```python
print("Below is the null values for each feature: ")
print(train.isna().sum())
```

Clearly there are no null values. Hence, there is no need to impute logically

3. Identify any categorical valued columns (non-numeric) and convert them to numeric.

```python
train['Furnishing Status']=train['Furnishing Status'].replace('Furnished',1)
train['Furnishing Status']=train['Furnishing Status'].replace('Semi-Furnished',2)
train['Furnishing Status']=train['Furnishing Status'].replace('Unfurnished',1)
```

4. Train a logistic regression model and check the score for different training iterations. Plot the
validation results by varying max_iter as 10, 20, 30, …. 

```python
test_train=train[['BHK','Size','Bathroom']]
logistic_model=LogisticRegression(solver='lbfgs',max_iter=10)
logistic_model.fit(test_train,train['Rent'])
```

```python
logistic_prediction=logistic_model.predict(test[['BHK','Size','Bathroom']])
```

```python
logistic_prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
mae = mean_absolute_error(test['Rent'],logistic_prediction)
mse = mean_squared_error(test['Rent'],logistic_prediction)
rmse = np.sqrt(mse)
rmse
```

```python
accuracy_score(test['Rent'],logistic_prediction)
```
