Split the training set into training and validation partitions in the ratio of 75% : 25% and train a linear 
classification model on the randomly shuffled training partition. Evaluate on the validation set. Make

```python
import pandas as pd
df=pd.read_csv('train.csv')
```

```python
df=df.sample(frac=1,random_state=42)
```

```python
df=df.drop(['PassengerId','Cabin','Ticket','Sex','Embarked','Name'],axis=1)
```

```python
df=df.dropna()
```

```python
df
```

```python
train=df.sample(frac=0.75,random_state=42)
test=df.drop(train.index)
```

```python
train
```

```python
validation=test
```

```python
validation
```

```python
y=train['Survived']
x=train.drop('Survived',axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
predictions = model.predict(test.drop('Survived',axis=1))
```

```python
print(classification_report(test['Survived'], predictions))
```
